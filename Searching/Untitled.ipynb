```python
#Binary search
def iterative_binary_search(arr,target):
    n=len(arr)
    low=0
    high=n-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]>target:
            high=mid-1
        else:
            low=mid+1
    return -1

def recursive_binary_search(arr, low, high, target):
    if low > high:
        return -1

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return recursive_binary_search(arr, low, mid - 1, target)
    else:
        return recursive_binary_search(arr, mid + 1, high, target)
```

```python
#Locate the first occurence of an element in an array
def first_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1  # Default if not found

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid  # Store it and keep searching left
            right = mid - 1
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result
    
```

```python
#Locate the last occurence of an element in an array
def last_occurence(arr,target):
    left=0
    right=len(arr)-1
    result=-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            result=mid
            left=mid+1
        elif arr[mid]<target:
            left=mid+1
        else:
            right=mid-1
    return result
```

```python
#Count occurences in sorted array
def first_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            right = mid - 1  # Move left to find earlier occurrence
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result

def last_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            left = mid + 1  # Move right to find later occurrence
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result

def count_occurrences(arr, target):
    first = first_occurrence(arr, target)
    last = last_occurrence(arr, target)
    if first == -1:
        return 0
    return last - first + 1
```

```python
#Count number of 1's in sorted array
def count_ones(arr):
    left = 0
    right = len(arr) - 1
    first_one = -1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == 1:
            first_one = mid
            right = mid - 1  # search left part
        else:
            left = mid + 1   # search right part
    
    if first_one == -1:
        return 0
    return len(arr) - first_one
```

```python
#Finding square root using binary search
def square_root(n):
    left = 0
    right = n
    ans = -1
    while left <= right:
        mid = (left + right) // 2
        if mid * mid == n:
            return mid
        elif mid * mid < n:
            ans = mid
            left = mid + 1
        else:
            right = mid - 1
    return ans
```

```python
def search_in_infinite_array(arr, target):
    left = 0
    right = 1

    # Expand the window
    while arr[right] < target:
        left = right
        right = right * 2

    # Normal binary search
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

arr = [3, 5, 7, 9, 10, 90, 100, 130, 140, 160, 170]
target = 10
print(search_in_infinite_array(arr, target))  # Output: 4
```

```python
#Search in sorted rotated array
def search_in_sorted_array(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        # Left half is sorted
        if arr[low] <= arr[mid]:
            if arr[low] <= target < arr[mid]:
                high = mid - 1
            else:
                low = mid + 1
        # Right half is sorted
        else:
            if arr[mid] < target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1
                
    return -1
```

```python
'''In this problem, we are given an array of unsorted integers . Our task is to find a peak element in the array.

A peak element is an element whose neighbors have value smaller than that of the element. '''
def find_peak(arr):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if (mid == 0 or arr[mid-1] <= arr[mid]) and (mid == len(arr)-1 or arr[mid] >= arr[mid+1]):
            return arr[mid]
        elif mid > 0 and arr[mid-1] > arr[mid]:
            right = mid - 1
        else:
            left = mid + 1
    return None
```

```python
#Two pointer approach for finding peak element
def find_peak(arr):
    left, right = 0, len(arr) - 1

    while left < right:
        mid = (left + right) // 2
        
        if arr[mid] < arr[mid + 1]:
            left = mid + 1
        else:
            right = mid
    
    return arr[left]  # left == right at the end
```

```python
def find_triplet(arr, target):
    n = len(arr)  # Step 1: Find the length of the array

    # Step 2: Loop through each element (fix the first element of triplet)
    for i in range(n-2):  # Only till n-3 because we need 3 elements
        left = i + 1      # Step 3: Set left pointer just after i
        right = n - 1     # Step 4: Set right pointer at the end of the array

        # Step 5: Use two pointers to find the other two elements
        while left < right:
            current_sum = arr[i] + arr[left] + arr[right]  # Sum of the triplet

            if current_sum == target:
                # Step 6: If sum matches target, return the triplet
                return (arr[i], arr[left], arr[right])

            elif current_sum < target:
                # Step 7: If current sum is smaller, move left pointer to right
                left += 1

            else:
                # Step 8: If current sum is larger, move right pointer to left
                right -= 1

    # Step 9: If no triplet found, return None
    return None
```
